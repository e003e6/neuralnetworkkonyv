# Aktivációs függvény

**Az aktiválási függvényt egy neuron (vagy neuronréteg) kimenetére alkalmazzuk, ami módosítja a kimeneteket.** A kiszámolt neuron output értéket beküldön a függvénybe és neuron valós kimenete a függvény Y értéke lesz.

Különböző esetekre különböző aktiválási függvényeket használunk.

Az aktiválási függvényeket azért használjuk, mert ***ha maga az aktiválási függvény nemlineáris, akkor (a két vagy több rejtett réteggel rendelkező) neurális hálózatok számára lehetővé teszi a nemlineáris függvények leképezését***.

A neurális hálózat kétféle aktiválási függvényt fog tartalmazni. Az első lesz a rejtett rétegekben használt aktiváló függvény, a második pedig a kimeneti rétegben használt.

## Gyakori aktivációs függvények


### Step activation funcion

Egy neuron "tüzelését" vagy "nem tüzelését" utánozza a bemeneti információ alapján. Ennek legegyszerűbb változata egy lépésfüggvény. Ha `súlyok*bemenetek+bias =` 0-nál nagyobb a neuron tüzel és 1-et ad ki, ellenkező esetben 0-t ad ki. Tehát a neuron végleges kimenete mindig 0 vagy 1.

$\large y=\left\{\begin{array}{ll}1 & x > 0 \\ 0 & x \leqslant 0\end{array}\right. $

Ezt az aktiválási függvényt régen használták a rejtett rétegekben, de manapság már ritkán használják.

```python
def step(x):
    if x > 0:
        return 1
    return 0
```

```python
printact(x, [step(i) for i in x])
```

<br>

### Linear activation function

A kimeneti érték megegyezik a bemeneti értékkel.

$\large y=x $

Ezt az aktiválási függvényt általában az utolsó réteg kimenetére alkalmazzák regressziós modell esetén (amikor a modellosztályozás helyett skalárértéket ad ki).

(Ha nem használunk aktivációs függvényt akkor lineáris aktivációs függvényt használunk.)

```python
printact(x, x, 2)
```

### Sigmoid activation funcion

**Azért hasznos mert az értéke azt adja meg milyen közel vagyunk ahhoz, hogy az érték 0 vagy éppen 1 legyen.** Az optimalizáló úgy működik, hogy értékeli a súlyok és torzítások egyéni hatásait a hálózat kimenetére. A lépésfüggvénnyel az a probléma, hogy az optimalizáló számára kevésbé egyértelmű, hogy mik ezek a hatások, mivel önmagában nagyon kevés információ: vagy be van kapcsolva (1) vagy ki van kapcsolva (0). Nehéz megmondani, hogy a bemeneti érték milyen "közel" volt az aktiváláshoz vagy a deaktiváláshoz.

Ez a függvény a negatív végtelentől 0-ig (tehát negatív bemenet esetén) 0-tól 0,5-ig terjedő értéket ad vissza, 0-tól végtelenig (tehát pozitív bemenet esetén) pedig 0-tól 1-ig terjedő értéket ad.

Vagyis egy olyan értéket kapunk, ami **visszafordítható az eredeti értékre**: a visszaadott érték tartalmazza az összes információt a bemenetről, ellentétben a lépésfüggvényel, ahol a 3 bemeneti érték ugyanazt az értéket fogja kiadni, mint a 300 000 bemeneti érték.

A sigmoid függvény kimenete, mivel a 0 és 1 közötti tartományban van, szintén jobban működik a neurális hálózatokkal (különösen a negatívtól a pozitív végtelenig terjedő tartományhoz képest), és nemlinearitást ad hozzá.

$\huge y=\frac{1}{1+e^{-x}} $

A rejtett rétegeken régen használt Sigmoid függvényt végül a Rectified Linear Units (vagy ReLU) aktiválási függvény váltotta fel.

```python
def sigmoid(x):
    return 1/(1+np.e**(-x))
```

```python
printact(x, [sigmoid(i) for i in x], 6)
```

<br>

### Rectified linear activation function

Az egyenesített lineáris aktiválási függvény egyszerűbb, mint a szigmoid. Ez szó szerint y=x, levágva 0-nál a negatív oldalról.

$\large y=\left\{\begin{array}{ll}x & x > 0 \\ 0 & x \leqslant 0\end{array}\right. $

Bár a szigmoid függvény nem túl bonyolult, mégis **sokkal nagyobb kihívást jelent a kiszámítása, mint a ReLU aktiválási függvényé**.

A ReLU aktiválási függvény rendkívül közel áll ahhoz, hogy lineáris aktiválási függvény legyen, miközben nemlineáris marad, köszönhetően  a 0 utáni vágásnak. Ez az egyszerű tulajdonság nagyon hatékony.

```python
def relu(x):
    if x>0:
        return x
    return 0
```

```python
printact(x, [relu(i) for i in x], 2)
```

## ReLU a gyakorlatban

Ahhoz, hogy egy neurális hálózat egy nemlineáris függvényhez illeszkedjen, két vagy több rejtett réteget kell tartalmaznia, és ezeknek a rejtett rétegeknek nemlineáris aktiválási függvényt kell használniuk.


A lineáris aktiválási függvénnyel egy 1 bemenetű és 1 kimenetű 2 rejtett rétegű, egyenként 8 neuronból álló neurális hálózat betanításának eredménye, valamint ha ugyanazt a 2 rejtett réteget használjuk, de ReLU aktivációs függvénnyel:

<img src="kepek/linearvsrelu.jpg" width="800"/>

Lineáris esetben bármit is teszünk a neuron súlyaival és torzításaival, a neuron kimenete lineáris lesz. Ez a lineáris jelleg az egész hálózatban fennmarad. Mivel minden egyes rétegben minden neuron lineárisan viselkedik, így **az egész hálózat is egy lineáris függvény**.

**De egy alig nemlineáris aktiváló függvénnyel, mint a ReLU nemlineáris függvényeket is le tudunk képezni.**


### ReLU működése

Az aktivációs függvény határozza meg, hogy a neuron tüzeljen e vagy sem. Végső soron pedig az érdekel minket **melyik neuronokat kapcsoljuk ki, mivel a végeredményt ez szabja meg**. Tehát szükség van rá, hogy legyen valami (egy függvény) amivel meg tudjuk határozni mikor működjön egy neuron.

**A súlyok határozzák meg a függvény meredekségét, a bias pedig a shifting-et, azaz hogy hol kezdődjön a lineáris függvény rész.** (Persze ez a felfogás csak egy absztrakt leképezés mivel nem az aktivációs függvény változik hanem a bemeneti érték, de a kettő felfogható felcserélve is.)

### Sok ReLU = nem lineáris függévény

Az alábbi ábrák azt mutatják, hogy **változó input adatokra milyen kimenetet ad a neuron és a ReLU függvény együttese**.

```python
def neouronOut(suly, bias, x):
    return x*suly+bias
```

```python
x = np.arange(-1, 1, 0.001)
y = [relu(neouronOut(1, 0.5, i)) for i in x]   # súly: 1 bias: 0.5
printrelu(x, y)
```

Láthatjuk, hogy a torzítás növelésével (0->0.5) a neuront korábban aktiváljuk, azaz eltorlhatjuk az akiválási pontját.

Mi történik, ha a súlyt -1-re állítjuk?

```python
x = np.arange(-1, 1, 0.001)
y = [relu(neouronOut(-1, 0.5, i)) for i in x]   # súly: 1 bias: 0.5
printrelu(x, y)
```

***Negatív súly esetén a funkció az lett, hogy deaktiválódik a neuron.***

Az aktivációs függvény segítségével a súlyokkal és a torzításokkal **azt tudjuk szabályozni, hogy a neuron hol aktiválódjon és hol deaktiválódjon.**

Ha 2 rejtett rétegünk van, egyenként 1 neuronból, azaz **a neuronokat sorba kapcsoljuk, akkor a kettő függvény együttese egy olyan függvényt ad aminek van egy aktiválási és egy deaktiválási pontja is**.

```python
x = np.arange(-1, 1, 0.001)
y = [relu(neouronOut(-2, 1, relu(neouronOut(-1, 0.5, i)))) for i in x]   # 1.neuron: súly: 1 bias: 0.5    2.neuron: súly: -2 bias: 1
printrelu(x, y)
```

Amikor mindkét neuron aktiválódik (amikor a "hatásterületük" működésbe lép), akkor a kimenet tartományába eső értékeket produkálnak. Ha a pár bármelyik neuronja inaktív, a pár nem változó kimenetet fog produkálni. **Ha több neuron párunk van akkor az a szakasz érvényes ahol mind két neuron aktiválódik**

## Softmax Activation Function

Ezt az aktiválási függvényt az utolsó réteg kimenetén használjuk.

A mi esetünkben ezt a modellt osztályozóvá szeretnénk tenni, tehát egy osztályozásra szánt aktiválási függvényt kell használnunk. Az egyik ilyen a Softmax aktiválási függvény.

A ReLU egység:
- nem normalizált: az értékek bármik lehetnek [0.05, 14, 318]
- kimenet kontextus nélküli
- kizárólagos: minden kimenet független a többitől

**A softmax függvény képes nem normalizált vagy kalibrálatlan bemeneteket fogadni, és az osztályaink valószínűségeinek normalizált eloszlását előállítani.**

Osztályozás esetén azt szeretnénk látni, hogy a hálózat "szerint" melyik osztályt képviseli a bemenet. Ez a softmax aktiválási függvény által visszaadott eloszlás az egyes osztályokra vonatkozó valószínűségi pontszámokat jelenti. Összege mindig 1. Például, ha a hálózatunknak két osztályra vonatkozó bizalmi eloszlása van: [0.45, 0.55], a jóslat a 2. osztály, de a bizalom nem túl magas. **Ez a függvény adja a jelentést az értékekhez.**

$$
\huge S_{i, j}=\frac{e^{z_{i, j}}}{\sum_{l=1}^L e^{z_{i, j}}}
$$


Az első lépés a kimenetek "exponenciálása". Ezt az Euler-számmal tesszük.

$\huge y = e^x$

A kimenet mindig pozitív, 
- ha a bemenet negatív, akkor nagyon kicsi szám
- ha a bemenet pozitív akkor hirtelen egyre nagyobb szám lesz

Erre azért van nagy szükség mert el akarjuk hagyni a negatív számokat, de a negatív számok jelentését nem. _(Ha abszolút értékre emeled, teljesen elveszted a negatív számok jelentését, ha négyzetre emeled torzulnak az értékek és jelentésük stb.)_

Az exponenciális függvény monoton függvény: a nagyobb bemeneti értékek esetén a kimenetek is magasabbak, így alkalmazás után nem fogjuk megváltoztatni a megjósolt osztályt, miközben nem negatív értékeket kapunk. Emellett stabilitást is ad az eredménynek, mivel a normalizált exponenciálás inkább a számok közötti különbségről szól, mint azok nagyságáról. 

```python
# három kimeneti neuronja van a hálóna, pl 3 osztályba szeretnénk besorolni valamit
layer_outputs = np.array([4.8, 1.21, 2.385]) 
```

```python
exp_values = np.e ** layer_outputs 
```

```python
exp_values
```

Ez után egy normalizálást fogunk végrehajtani, ahol egy adott értéket veszünk, és elosztjuk az összes érték összegével.

Két kimenet esetén:

$\huge y=\left[\begin{array}{c}\frac{u_{1}}{u_{1}+u_{2}} \\ \frac{u_{2}}{u_{1}+u_{2}}\end{array}\right]=\left[\begin{array}{c}\frac{1}{1+2} \\ \frac{2}{1+2}\end{array}\right]=\left[\begin{array}{l}\frac{1}{3} \\ \frac{2}{3}\end{array}\right]=\left[\begin{array}{l}0.33 \\ 0.67\end{array}\right]$

```python
exp_values/np.sum(exp_values)
```

És meg is kaptuk a valószínűségeket (összegük 1)

### Batch

A következő lépés, hogy egyszerre nem egy inputot számoltatunk végig a hálón hanem Batch-t,

```python
layer_outputs = np.array([[4.8, 1.21, 2.385], 
                          [8.9, -1.81, 0.2], 
                          [1.41, 1.051, 0.026]]) 
```

A kérdés az, hogyan kell vektorosan osztani a sorok összegével

```python
# összeadom a sorokat:
np.sum(layer_outputs, axis=1, keepdims=True) # keepdims -> megtartja az eredeti dimenzó típusát
```

```python
exp_values = np.exp(layer_outputs)
```

```python
exp_values / np.sum(exp_values, axis=1, keepdims=True)
```

És itt jön képbe a túlfolyás probléma: ha túl nagy a bemenet akkor túl sok lesz a tizedes jegy ami erőforrásokat pazarol.

Ezt úgy oldjuk meg, hogy (overflow prevention) minden kimeneti értékből kivonom a legnagyobb kimeneti értéket, így a legnagyobb érték nulla lesz minden pás eddig ennél kissebb. (Persze itt is minden sorokra, azaz bemeneti adatokra vonatkozik.)

$\large v = u-max u$

Ezt a lépést az exponálás elött kell végrehajtani. (Ezzel garantáljuk, az exponálás eredménye 0 és 1 között maradjon)

```python
v = layer_outputs - np.max(layer_outputs, axis=1, keepdims=True)
exp_values = np.exp(v)
norm_values = exp_values / np.sum(exp_values, axis=1, keepdims=True)
```

```python
norm_values
```

```python
class Activation_Softmax:
    def forward(self, inputs):
        v = inputs - np.max(inputs, axis=1, keepdims=True)
        exp_values = np.exp(v)
        norm_values = exp_values / np.sum(exp_values, axis=1, keepdims=True) # norm_values = probabilities
        self.probabilities = norm_values
```
